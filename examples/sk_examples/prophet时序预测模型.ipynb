## 这里演示的是prophet时序模型，其原理是将季节性等较为现实的因素考虑进去

其公式如下 $$prophet = x_{(t)} + y_{(t)} + z_{(t)} + \alpha $$
其中$ x_{(t)}$代表数据的整体变动趋势（增加或者下降）;$ y_{(t)}$代表季节性的影响; $ z_{(t)}$代表节假日的影响。

适用：
- 电商
- 随着季节性变动较为明显
- 具有实际意义的时序数据

不适用
- 股票等随机性太大的数据
- 受外界影响较大的数据

这边调用prophet库进行实现
